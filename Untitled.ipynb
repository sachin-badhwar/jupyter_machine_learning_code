```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some linear-looking data
np.random.seed(42)  # for reproducibility
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1) # y = 4 + 3x + noise

from sklearn.linear_model import LinearRegression

# Create a Linear Regression model and fit it to the data
model = LinearRegression()
model.fit(X, y)

# Make predictions on the same data
y_predicted = model.predict(X)

# Calculate the residuals
residuals = y - y_predicted

# Create the residual plot
plt.figure(figsize=(10, 6))
plt.scatter(y_predicted, residuals)

# Add a horizontal line at y=0 for reference
plt.hlines(y=0, xmin=y_predicted.min(), xmax=y_predicted.max(), color='red', linestyle='--')

# Add labels and title
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot of a Linear Model on Linear Data")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Generate non-linear data (e.g., a quadratic relationship)
X_nonlinear = 4 * np.random.rand(100, 1) - 2
y_nonlinear = 2 + 3 * X_nonlinear + 0.5 * X_nonlinear**2 + np.random.randn(100, 1)

# Fit a linear model to the non-linear data
model_nonlinear = LinearRegression()
model_nonlinear.fit(X_nonlinear, y_nonlinear)

y_predicted_nonlinear = model_nonlinear.predict(X_nonlinear)

# Calculate and plot residuals for the non-linear data
residuals_nonlinear = y_nonlinear - y_predicted_nonlinear

plt.figure(figsize=(10, 6))

plt.scatter(y_predicted_nonlinear, residuals_nonlinear)
plt.hlines(y=0, xmin=y_predicted_nonlinear.min(), xmax=y_predicted_nonlinear.max(), color='red', linestyle='--')
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot of a Linear Model on Non-Linear Data")
plt.grid(True)
plt.show()
```
